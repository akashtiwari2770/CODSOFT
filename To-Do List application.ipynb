```python
import datetime
```

```python
def add_task():
    print("\nAdding a task")
    task_name = input("Enter the name of the task: ")
    task_description = input("Enter the description of the task: ")
    task_due_date = input("Enter the due date of the task (DD/MM/YYYY): ")

    tasks[task_name] = {"Name": task_name, "Description": task_description, "Due Date": task_due_date}
    print("\nTask added successfully!")

def view_tasks():
    print("\nViewing tasks")

    if len(tasks) == 0:
        print("No tasks to view!")
    else:
        for task in tasks.values():
            print(f"Name: {task['Name']}\nDescription: {task['Description']}\nDue Date: {task['Due Date']}\n")

def update_task():
    print("\nUpdating a task")
    task_name = input("Enter the name of the task to update: ")

    if task_name in tasks:
        task_description = input("Enter the updated description of the task: ")
        task_due_date = input("Enter the updated due date of the task (DD/MM/YYYY): ")

        tasks[task_name] = {"Name": task_name, "Description": task_description, "Due Date": task_due_date}
        print("\nTask updated successfully!")
    else:
        print("\nTask not found!")

def delete_task():
    print("\nDeleting a task")
    task_name = input("Enter the name of the task to delete: ")

    if task_name in tasks:
        del tasks[task_name]
        print("\nTask deleted successfully!")
    else:
        print("\nTask not found!")
def main():
    while True:
        print("\n1. Add Task")
        print("2. View Tasks")
        print("3. Update Task")
        print("4. Delete Task")
        print("5. Exit")

        option = input("Enter your option: ")

        if option == "1":
            add_task()
        elif option == "2":
            view_tasks()
        elif option == "3":
            update_task()
        elif option == "4":
            delete_task()
        elif option == "5":
            print("\nExiting the program...")
            break
        else:
            print("\nInvalid option! Please try again.")

if __name__ == "__main__":
    tasks = {}
    main()
```
